IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

IMPORTING DATA

```python
data = pd.read_csv("Data - Data.csv")
data
```

CREATING THE YOY CHANGE DATA TO FIT INTO MODEL

```python
data_array = np.array(data)
new_arr = []
for i in range(len(data_array)-1):
    
    t_1 = data_array[i]
    t = data_array[i+1]

    year = f'{t[0]} - {t_1[0]}'

    gdp = float(t[1]) - float(t_1[1])
    revenue = float(t[2]) - float(t_1[2])
    capital = float(t[3]) - float(t_1[3])
    total = float(t[4]) - float(t_1[4])

    new_arr.append([year, gdp, revenue, capital, total])

yoy_change_data = pd.DataFrame(new_arr, columns=["Period", "GDP", "Revenue expenditure", "Capital expenditure", "Total expenditure"])
yoy_change_data = yoy_change_data.drop(36) #Dropping the last observation since economy was hit by covid, a black swan event and there was no data ahead of it to bother with dummy variable
yoy_change_data
```

TAKING COMPONENTS FOR GDP, REVENUE&CAPITAL EXPENDITURE AND TOTAL EXPENDITURE TO FIT INTO MODELS

```python
#GDP
yGDP = np.array(yoy_change_data.iloc[:, 1]).reshape(-1,1)

#Revenue and Capital expenditure
xExpenditure = np.array(yoy_change_data.iloc[:, 2:4])

#Total expenditure
xTotal_expenditure = np.array(yoy_change_data.iloc[:, -1]).reshape(-1,1)
```

CREATIND MODEL TAKING REVENUE AND CAPITAL EXPENDITURE       
Y = a + b*(revenue expenditure) + c*(capital expenditure)

```python
#Creating Model
rev_cap_model = LinearRegression()

#Fitting the data into it
rev_cap_model.fit(xExpenditure, yGDP)

print("Accuracy score = ",rev_cap_model.score(xExpenditure, yGDP))
print("Coefficients = ",rev_cap_model.coef_)
```

CREATING MODEL TAKING TOTAL EXPENDITURE             
Y = a + b*(total expenditure)

```python
#creating model
Total_exp_model = LinearRegression()

#Fitting data into it
Total_exp_model.fit(xTotal_expenditure, yGDP)

print("Accuracy score = ",Total_exp_model.score(xTotal_expenditure, yGDP))
print("Coefficients = ",Total_exp_model.coef_)
```

SINCE INDIAN ECONOMY HAD GONE THROUGH MASSIVE REFORM IN 1991 AS IT OPENED ITS WISER TO ADD A DUMMY SEPERATING THE TWO ERAS AS GDP GROWTH WAS MUCH HIGHER AFTERWARDS

```python
#Creating a new dataframe
new_change_data = pd.DataFrame(yoy_change_data, columns=yoy_change_data.columns)

#Adding dummy variable
new_change_data["Dummy 1"] = 0
new_change_data["Dummy 1"] = new_change_data["Dummy 1"].mask(new_change_data.index > 6, 1)

new_change_data
```

CREATING A NEWER MODEL USING THIS DUMMY VARIABLE ALONG WITH THE TOTAL EXPENDITURE                   
Y = a + b*(Total expenditure) + c*(Dummy variable)

```python
#Taking both total expenditure and dummy variable for predicting variable
x2Total_expenditure = new_change_data.iloc[:, -2:]

#Creating model
newTotal_expenditure_model = LinearRegression()

#Fitting data into it
newTotal_expenditure_model.fit(np.array(x2Total_expenditure), yGDP)

print("Accuracy score : ", newTotal_expenditure_model.score(np.array(x2Total_expenditure), yGDP))
print("Coeffecients = ", newTotal_expenditure_model.coef_)
```

ADDING THE DUMMY VARIABLE INCREASED THE ACCURACY, NOW WE WILL ADD ANOTHER DUMMY VARIABLE FORMED BY (Total Expenditure)*(Dummy 1) TO SEE IF WE CAN GET EVEN BETTER RESULTS

```python
new_change_data["Dummy 2"] = new_change_data["Total expenditure"]*new_change_data["Dummy 1"]
new_change_data
```

CREATING ANOTHER MODEL USING BOTH DUMMIES AND TOTAL EXPENDITURE                   
Y = a + b*(Total Expenditure) + c*(Dummy 1) + d*(Dummy 2)

```python
#Taking total expenditure and the two dummy variables
x3Total_expenditure = new_change_data.iloc[:, -3:]

#Creating model
nextTotal_expenditure_model = LinearRegression()

#Fitting data into it
nextTotal_expenditure_model.fit(np.array(x3Total_expenditure), yGDP)

print("Accuracy score : ", nextTotal_expenditure_model.score(np.array(x3Total_expenditure), yGDP))
print("Coeffecients = ", nextTotal_expenditure_model.coef_)
```

NOW WE HAVE TO CHECK WHETHER ACCURACY INCREASE IN THE NEW MODEL IS STATISTICALLY SIGNIFICANT AND WORTH ADDING AN EXTRA VARIABLE                     

TO DO THIS WE WILL CONDUCT THE PARTIAL F TEST FOR COMPARING TWO LR MODELS               

F = ((SS1 - SS2)/(df1 - df2))/(SS2/df2)                 
SS : Sum of sqaured errors                                  
df : Degree of freedom                  
1 : Reduced model               
2 : Full model

Step 1 : Calculating the sum of sqaured errors for both models

```python
new_SSE = 0
for i in range(len(x2Total_expenditure)):
    new_SSE += (newTotal_expenditure_model.predict([[x2Total_expenditure.iloc[i,0], x2Total_expenditure.iloc[i,1]]]) - yGDP[i])**2

next_SSE = 0
for i in range(len(x3Total_expenditure)):
    next_SSE += (nextTotal_expenditure_model.predict([[x3Total_expenditure.iloc[i,0], x3Total_expenditure.iloc[i,1], x3Total_expenditure.iloc[i,2]]]) - yGDP[i])**2

print(new_SSE - next_SSE)
```

Step 2: Calculating the mean squared error for the full model(nextTotal_expenditure_model) and finding F value

```python
next_MSE = next_SSE/len(x3Total_expenditure)
f = (new_SSE - next_SSE)/next_MSE
f
```

The calculated F statistic is lower than the critical F and hence null hypothesis cannot be rejected so the reduced model(newTotal_expenditure_model) is better than full model(nextTotal_expenditure_model)
